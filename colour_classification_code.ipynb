```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#part a:

# Replace 'your_filename.csv' with the actual path and name of your CSV file
file_path = '/content/Colour classification of galaxies (part a).csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Now 'df' contains the data from the CSV file
print(df.head())
```

```python
# Slice the DataFrame by selecting specific rows (0 to 9) and columns ('u-g' and 'g-r')
sliced_u_g = df.loc[0:9, ['u-g']]
sliced_g_r = df.loc[0:9, ['g-r']]

# Now 'sliced_df' contains the sliced data
print(sliced_u_g,)
print(sliced_g_r)

plt.plot(sliced_u_g, sliced_g_r, 'r.',label='data')
plt.xlabel('(u-g)')
plt.ylabel('(g-r)')
plt.title('colour colour diagram / (u-g) vs (g-r) plot')
plt.legend()
plt.grid ()

x_values = np.linspace(-5, 5, 100)  # You can adjust the range and number of points

# Calculate corresponding y values for the line y = -x + 2.2
y_values = -x_values + 2.2

# Plot the line
plt.plot(x_values, y_values, label='y = -x + 2.2')

plt.legend()

# Set the x-axis and y-axis limits
plt.xlim([1, 2.5])
plt.ylim([0, 1.2])

print("there are 2 data points at (1.35,0.62). above is spiral, below is eliptical.")
```

```python
#part b:

# Replace 'your_filename.csv' with the actual path and name of your CSV file
file_path = '/content/part b galaxy data.csv'

# Read the CSV file into a DataFrame
datab = pd.read_csv(file_path)

# Now 'df' contains the data from the CSV file
print(datab.head())
```

```python
# Slice the DataFrame by selecting specific rows (0 to 9) and columns ('u-g' and 'g-r')
sliced_u = datab['u']
sliced_g = datab['g']
sliced_r = datab['r']


ug = sliced_u - sliced_g
gr = sliced_g - sliced_r

plt.plot(ug, gr, 'r.',label='data')
plt.xlabel('(u-g)')
plt.ylabel('(g-r)')
plt.title('colour colour diagram / (u-g) vs (g-r) plot')
plt.legend()
plt.grid ()
```

```python
# Replace 'your_filename.csv' with the actual path and name of your CSV file
file_path = '/content/Skyserver_Radial Abell 2255.csv'

# Read the CSV file into a DataFrame
datab2 = pd.read_csv(file_path)

# Now 'df' contains the data from the CSV file
print(datab2.head())
```

```python
# Slice the DataFrame by selecting specific rows (0 to 9) and columns ('u-g' and 'g-r')

u = datab2['u']
g = datab2['g']
r = datab2['r']
ugb2 = datab2['u-g']
grb2 = datab2['g-r']
ur = u - r

# Now 'sliced_df' contains the sliced data
print(ugb2)
print(grb2)


plt.plot(ugb2, grb2, 'r.',label='data')
plt.xlabel('(u-g)')
plt.ylabel('(g-r)')
plt.title('colour colour diagram / (u-g) vs (g-r) plot')
plt.legend()
plt.grid ()

x_values = np.linspace(-5, 8, 100)  # You can adjust the range and number of points

# Calculate corresponding y values for the line y = -x + 2.2
y_values = -x_values + 2.2

# Plot the line
plt.plot(x_values, y_values, label='y = -x + 2.2')

plt.plot(ug, gr, 'b.',label='20 glaxies')

# Set the x-axis and y-axis limits
plt.xlim([-5, 8])
plt.ylim([-6.5, 10.5])

plt.legend()
```

```python
y = grb2
x = ugb2

num_s = len(y[y>-x+2.22])
num_e = len(y[y<-x+2.22])

print(f"there are {num_s} spirtals")
print(f"there are {num_e} elipticals")
```

```python

g_magrange = len(g[(g > 15) & (g < 19)])
g_magrange2 = len(g[(g > 19) & (g < 23)])

print(f"g magnitude between 15 and 19 = {g_magrange}")
print(f'g magnitude between 19 and 23 = {g_magrange2}')
```

```python
ur_1 = ur[(g > 15) & (g < 19)]
ur_2 = ur[(g > 19) & (g < 23)]

g_1 = g[(g > 15) & (g < 19)]
g_2 = g[(g > 19) & (g < 23)]

print(f" mean ur when g is between 15 and 19 = {np.mean(ur[(g > 15) & (g < 19)])}")
print(f"mean ur when g is between 19 and 23 = {np.mean(ur[(g > 19) & (g < 23)])}")
```

```python
plt.plot(ur_1, g_1, 'r.',label='data')
plt.plot(ur_2, g_2, 'b.',label='data')
plt.xlabel('(u-r)')
plt.ylabel('g')
plt.title('colour colour diagram / (u-r) vs g plot')
plt.legend()
plt.grid ()
```
